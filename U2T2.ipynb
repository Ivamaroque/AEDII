### Objetivo:
**Montar um gráfico de assortividade para cada rede escolhida**

```python
# Bibliotecas
import networkx as nx
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Baixando os datasets das 5 redes escolhidas
!wget https://snap.stanford.edu/data/web-BerkStan.txt.gz
!wget https://snap.stanford.edu/data/email-EuAll.txt.gz
!wget https://snap.stanford.edu/data/web-Google.txt.gz
!wget https://snap.stanford.edu/data/wiki-RfA.txt.gz
!wget https://snap.stanford.edu/data/amazon0601.txt.gz
```

```python
web_Berkstan= nx.read_edgelist('web-BerkStan.txt.gz', create_using=nx.DiGraph(), data=False)
email_EuAll = nx.read_edgelist('email-EuAll.txt.gz', create_using=nx.DiGraph(), data=False)
web_Google = nx.read_edgelist('web-Google.txt.gz', create_using=nx.Graph(), data=False)
wiki_RfA = nx.read_edgelist('wiki-RfA.txt.gz', create_using=nx.DiGraph(), data=False)
amazon0601 = nx.read_edgelist('amazon0601.txt.gz', create_using=nx.DiGraph(), data=False)
```

### Criando gráfico para web_Berkstan


```python
# Calcula a média dos graus dos vizinhos
average_degree_connectivity = nx.average_degree_connectivity(web_Berkstan)

# Extraia as informações em listas separadas
degrees, avg_neighbor_degrees = zip(*average_degree_connectivity.items())

# Converta as tuplas para listas
degrees = list(degrees)
avg_neighbor_degrees = list(avg_neighbor_degrees)

# Configurar o estilo do gráfico
plt.style.use("fivethirtyeight")

# Criar uma figura e um eixo para o gráfico
fig, ax = plt.subplots(figsize=(12, 8))

# Criar o gráfico de dispersão (scatter plot) com regressão linear
sns.regplot(x=degrees, y=avg_neighbor_degrees, ax=ax)

# Defina rótulos dos eixos
ax.set_xlabel("Grau do Nodo")
ax.set_ylabel("Grau Médio dos Vizinhos")
ax.set_xlim(0, 3000)

# Adicione um título ao gráfico
ax.set_title("Gráfico de Assortatividade da Web Berkstan")

# Exibir o gráfico
plt.show()
```
